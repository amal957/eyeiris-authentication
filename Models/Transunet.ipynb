```python
import numpy as np
import cv2
import os
import tensorflow as tf
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Input, concatenate, Dropout
from tensorflow.keras.layers import LayerNormalization, Dense, MultiHeadAttention, Add
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Set the plotting style globally
plt.style.use('seaborn-darkgrid')

# Step 1: Data Loading
def load_images_and_masks(dataset_dir, target_size=(128, 128)):
    images, masks = [], []
    
    print(f"Starting data loading from: {dataset_dir}")
    
    # Loop through folders 001 to 223
    for folder_num in range(1, 224):
        folder_name = f"{folder_num:03d}"  # Format as 001, 002, etc.
        folder_path = os.path.join(dataset_dir, folder_name)
        
        # Check if folder exists
        if not os.path.exists(folder_path):
            print(f"Skipping folder {folder_name} - does not exist at {folder_path}")
            continue
        
        try:
            # Get all image files (.bmp) in the current folder
            image_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.bmp')]
            
            if not image_files:
                print(f"No .bmp files found in folder {folder_name}")
                continue
                
            print(f"Processing folder {folder_name} - Found {len(image_files)} images")
            
            for image_file in image_files:
                try:
                    # Load image (.bmp format)
                    image_path = os.path.join(folder_path, image_file)
                    image = cv2.imread(image_path)
                    
                    # Load corresponding mask (.png format)
                    mask_file = image_file.replace('.bmp', '.png')
                    mask_path = os.path.join(folder_path, mask_file)
                    
                    if not os.path.exists(mask_path):
                        print(f"Missing mask file for {image_file} in {folder_name}")
                        continue
                        
                    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
                    
                    if image is None:
                        print(f"Failed to load image: {image_path}")
                        continue
                    if mask is None:
                        print(f"Failed to load mask: {mask_path}")
                        continue
                    
                    # Resize images and masks
                    image = cv2.resize(image, target_size)
                    mask = cv2.resize(mask, target_size)
                    
                    # Normalize the data
                    images.append(img_to_array(image) / 255.0)
                    masks.append(img_to_array(mask) / 255.0)
                    
                    if len(images) % 10 == 0:
                        print(f"Processed {len(images)} images...")
                        
                except Exception as e:
                    print(f"Error processing {image_file} in {folder_name}: {str(e)}")
                    continue
                    
        except Exception as e:
            print(f"Error processing folder {folder_name}: {str(e)}")
            continue
    
    if not images:
        raise ValueError("No images were successfully loaded!")
    
    # Convert to numpy arrays
    images_array = np.array(images)
    masks_array = np.array(masks)
    
    print(f"\nFinished loading:")
    print(f"Total images loaded: {len(images)}")
    print(f"Images shape: {images_array.shape}")
    print(f"Masks shape: {masks_array.shape}")
    
    return images_array, masks_array

def swin_transformer_block(input_tensor, num_heads=8, head_dim=32):
    """
    Keras-compatible Swin Transformer block implementation
    """
    # Multi-head self attention
    attention = MultiHeadAttention(
        num_heads=num_heads,
        key_dim=head_dim,
        value_dim=head_dim
    )(input_tensor, input_tensor)
    
    # Add skip connection and normalization
    attention = LayerNormalization()(Add()([input_tensor, attention]))
    
    # MLP block
    channels = input_tensor.shape[-1]
    mlp = Dense(channels * 4, activation='gelu')(attention)
    mlp = Dense(channels)(mlp)
    mlp = Dropout(0.1)(mlp)
    
    # Final skip connection and normalization
    output = LayerNormalization()(Add()([attention, mlp]))
    
    return output

def TransUNet(input_shape=(128, 128, 3), num_classes=1):
    inputs = Input(input_shape)

    # Encoder
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
    pool1 = Dropout(0.1)(pool1)

    conv2 = Conv2D(128, 3, activation='relu', padding='same')(pool1)
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
    pool2 = Dropout(0.1)(pool2)

    conv3 = Conv2D(256, 3, activation='relu', padding='same')(pool2)
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(conv3)
    pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)
    pool3 = Dropout(0.2)(pool3)

    # Bottleneck with Swin Transformer
    bottleneck = swin_transformer_block(pool3, num_heads=8, head_dim=32)

    # Decoder
    up6 = UpSampling2D(size=(2, 2))(bottleneck)
    up6 = Conv2D(256, 2, activation='relu', padding='same')(up6)
    merge6 = concatenate([conv3, up6], axis=3)
    conv6 = Conv2D(256, 3, activation='relu', padding='same')(merge6)
    conv6 = Conv2D(256, 3, activation='relu', padding='same')(conv6)
    conv6 = Dropout(0.2)(conv6)

    up7 = UpSampling2D(size=(2, 2))(conv6)
    up7 = Conv2D(128, 2, activation='relu', padding='same')(up7)
    merge7 = concatenate([conv2, up7], axis=3)
    conv7 = Conv2D(128, 3, activation='relu', padding='same')(merge7)
    conv7 = Conv2D(128, 3, activation='relu', padding='same')(conv7)
    conv7 = Dropout(0.1)(conv7)

    up8 = UpSampling2D(size=(2, 2))(conv7)
    up8 = Conv2D(64, 2, activation='relu', padding='same')(up8)
    merge8 = concatenate([conv1, up8], axis=3)
    conv8 = Conv2D(64, 3, activation='relu', padding='same')(merge8)
    conv8 = Conv2D(64, 3, activation='relu', padding='same')(conv8)
    conv8 = Dropout(0.1)(conv8)

    conv9 = Conv2D(num_classes, 1, activation='sigmoid')(conv8)

    model = Model(inputs, conv9)
    return model

# Metrics
def dice_coefficient(y_true, y_pred, smooth=1):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)
def combined_loss(y_true, y_pred):
    """
    Combines dice loss and binary crossentropy
    """
    dice = dice_loss(y_true, y_pred)
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    return dice * 0.5 + bce * 0.5
def dice_loss(y_true, y_pred):
    return 1 - dice_coefficient(y_true, y_pred)

def iou(y_true, y_pred, smooth=1):
    y_true = tf.cast(y_true > 0.5, tf.float32)
    y_pred = tf.cast(y_pred > 0.5, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) - intersection
    return (intersection + smooth) / (union + smooth)

# Training function
def train_model(model, X_train, y_train, X_val, y_val, epochs=70, batch_size=8):
    # Compile model with single combined loss
    model.compile(
        optimizer=Adam(learning_rate=1e-4),
        loss=combined_loss,  # Using single combined loss instead of list
        metrics=['accuracy', dice_coefficient, iou]
    )
    
    # Learning rate scheduler
    def lr_schedule(epoch):
        initial_lr = 1e-4
        if epoch < 30:
            return initial_lr
        elif epoch < 50:
            return initial_lr * 0.1
        else:
            return initial_lr * 0.01
    
    callbacks = [
        # Learning rate scheduler
        tf.keras.callbacks.LearningRateScheduler(lr_schedule),
        
        # Model checkpointing
        tf.keras.callbacks.ModelCheckpoint(
            'best_model.keras',
            monitor='val_dice_coefficient',
            mode='max',
            save_best_only=True,
            verbose=1
        ),
        
        # Early stopping
        tf.keras.callbacks.EarlyStopping(
            monitor='val_dice_coefficient',
            mode='max',
            patience=10,
            restore_best_weights=True,
            verbose=1
        ),
        
        # TensorBoard logging
        tf.keras.callbacks.TensorBoard(
            log_dir='./logs',
            histogram_freq=1,
            write_graph=True
        )
    ]

    # Train the model
    print("Starting model training...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=callbacks,
        verbose=1
    )
    
    return model, history

# Visualization functions
def visualize_results(model, X_val, y_val, num_samples=5):
    for i in np.random.randint(0, len(X_val), num_samples):
        original_image = X_val[i]
        ground_truth_mask = y_val[i].squeeze()
        predicted_mask = model.predict(np.expand_dims(original_image, axis=0)).squeeze()
        
        iou_score = iou(ground_truth_mask, predicted_mask).numpy()
        dice_score = dice_coefficient(ground_truth_mask, predicted_mask).numpy()
        
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        fig.suptitle(f'Sample {i+1} - IoU: {iou_score:.2f}, Dice: {dice_score:.2f}', fontsize=14)
        
        axes[0].imshow(original_image)
        axes[0].set_title('Original Image')
        axes[0].axis('off')
        
        axes[1].imshow(ground_truth_mask, cmap='gray')
        axes[1].set_title('Ground Truth Mask')
        axes[1].axis('off')
        
        im = axes[2].imshow(predicted_mask > 0.5, cmap='gray')
        axes[2].set_title('Predicted Mask')
        axes[2].axis('off')
        
        plt.colorbar(im, ax=axes[2], fraction=0.046, pad=0.04)
        plt.tight_layout()
        plt.show()

def plot_training_history(history):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    ax1.plot(history.history['loss'], label='Training Loss')
    ax1.plot(history.history['val_loss'], label='Validation Loss')
    ax1.set_title('Model Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend(loc='upper right')
    
    ax2.plot(history.history['accuracy'], label='Training Accuracy')
    ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax2.set_title('Model Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend(loc='lower right')
    
    ax3.plot(history.history['dice_coefficient'], label='Training Dice')
    ax3.plot(history.history['val_dice_coefficient'], label='Validation Dice')
    ax3.set_title('Dice Coefficient')
    ax3.set_xlabel('Epoch')
    ax3.set_ylabel('Dice Coefficient')
    ax3.legend(loc='lower right')
    
    ax4.plot(history.history['iou'], label='Training IoU')
    ax4.plot(history.history['val_iou'], label='Validation IoU')
    ax4.set_title('IoU Score')
    ax4.set_xlabel('Epoch')
    ax4.set_ylabel('IoU')
    ax4.legend(loc='lower right')
    
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    # Set random seed for reproducibility
    tf.random.set_seed(42)
    np.random.seed(42)
    
    # Step 1: Load and preprocess data
    dataset_dir = "/kaggle/input/iitd-iris/Masks and Images"  # Update with your dataset path
    images, masks = load_images_and_masks(dataset_dir)
    print("Data loaded. Images shape:", images.shape, "Masks shape:", masks.shape)
    
    # Step 2: Split data
    X_train, X_val, y_train, y_val = train_test_split(
        images, masks, 
        test_size=0.2, 
        random_state=42
    )
    print("Data split. Training set:", X_train.shape, y_train.shape)
    print("Validation set:", X_val.shape, y_val.shape)
    
    # Step 3: Create and compile model
    model = TransUNet(input_shape=(128, 128, 3), num_classes=1)
    model.summary()
    
    # Step 4: Train model
    trained_model, history = train_model(
        model=model,
        X_train=X_train,
        y_train=y_train,
        X_val=X_val,
        y_val=y_val,
        epochs=70,
        batch_size=8
    )
    
    # Step 5: Save the trained model
    print("\nSaving final model...")
    trained_model.save('final_model.keras')  # Changed from .h5 to .keras
    print("Model saved as 'final_model.keras'")
    
    # Save training history
    print("Saving training history...")
    np.save('training_history.npy', history.history)
    print("Training history saved as 'training_history.npy'")
    
    # Step 6: Visualize results
    print("\nGenerating visualization of results...")
    visualize_results(trained_model, X_val, y_val, num_samples=5)
    
    # Step 7: Plot training history
    print("\nPlotting training history...")
    plot_training_history(history)
    
    # Step 8: Final evaluation
    print("\nEvaluating final model on validation set...")
    evaluation = trained_model.evaluate(
        X_val, 
        y_val, 
        verbose=1
    )
    
    # Print final metrics
    print("\nFinal Evaluation Metrics:")
    metrics = ['Loss', 'Accuracy', 'Dice Coefficient', 'IoU']
    for metric, value in zip(metrics, evaluation):
        print(f"{metric}: {value:.4f}")

# To load and use the saved model later:
def load_saved_model(model_path='final_model.keras'):  # Changed from .h5 to .keras
    """
    Load a saved model with custom metrics
    """
    return tf.keras.models.load_model(
        model_path,
        custom_objects={
            'dice_coefficient': dice_coefficient,
            'dice_loss': dice_loss,
            'iou': iou
        }
    )

# Example of using the loaded model:
def predict_with_saved_model(image_path, model_path='final_model.keras'):  # Changed from .h5 to .keras
    """
    Make a prediction using a saved model
    """
    # Load the model
    model = load_saved_model(model_path)
    
    # Load and preprocess the image
    image = cv2.imread(image_path)
    image = cv2.resize(image, (128, 128))
    image = img_to_array(image) / 255.0
    image = np.expand_dims(image, axis=0)
    
    # Make prediction
    prediction = model.predict(image)
    
    return prediction
```
